```python
import nltk
import pandas as pd
from nltk.corpus import stopwords
import re
from bs4 import BeautifulSoup
from spellchecker import SpellChecker
from sklearn.feature_extraction.text import CountVectorizer
import spacy
import numpy as np
```

```python
from transformers import pipeline

# Initialize the text generation pipeline with the desired model
generator = pipeline('text-generation', model='gpt2')
```

```python
# Generate first document using the model
doc1 = generator('Olympics', max_length=1000, num_return_sequences=1,truncation=True)
doc1 = doc1[0]['generated_text']
# Generate second document using the model
doc2 = generator('Corona Virus', max_length=1000, num_return_sequences=1,truncation=True)
doc2 = doc2[0]['generated_text']
# Generate third document using the model
doc3 = generator('Egypt Pyramids', max_length=1000, num_return_sequences=1,truncation=True)
doc3 = doc3[0]['generated_text']
```

```python
# Print three documents
print(doc1)
print('------------------------------------------------------------------------------------------------------------------------')
print(doc2)
print('------------------------------------------------------------------------------------------------------------------------')
print(doc3)
```

```python
def cleaning(data):

    nlp = spacy.load('en_core_web_sm')
    # Process the text using spaCy
    data = nlp(data)
    # Extract lemmatized tokens
    lemmatized_tokens = [token.lemma_ for token in data]
    # Join the lemmatized tokens into a sentence
    data = ' '.join(lemmatized_tokens)
    data = data.lower()
    # Remove html tags
    data = BeautifulSoup(data, "html.parser").get_text()
    # Remove links and emails
    data = re.sub(r'http\S+|www\S+|[\w\.-]+@[\w\.-]+', '', data)
    # Remove any thing except words, numbers, and space
    data = re.sub(r'[^\w\s]', '', data)
    # Remove numbers
    data = re.sub(r'[0-9]', '', data)
    # Remove emojis and non ascii characters
    data = re.sub(r'[^\x00-\x7F]+', '', data)
    # Correct words
    spell = SpellChecker()
    data = ' '.join([spell.correction(word) if spell.correction(word) is not None else "" for word in data.split()])
    # Remove stop words (the, and, ....)
    stop_words = set(stopwords.words("english"))
    data = ' '.join([word for word in data.split() if word not in stop_words])

    # Split sentences to words
    data = data.split()
    data = ' '.join([word for word in data if (len(word)>1 or word == 'i')])
    return data
```

```python
doc1_cleaned = cleaning(doc1)
doc2_cleaned = cleaning(doc2)
doc3_cleaned = cleaning(doc3)
print(doc1_cleaned)
print(doc2_cleaned)
print(doc3_cleaned)
```

```python
corpus = [doc1_cleaned, doc2_cleaned, doc3_cleaned]
```

```python
all_words = set()
for doc in corpus :
  all_words = all_words.union(set(doc.split()))
print('Number of words:',len(all_words))
print('Unique words: \n', all_words)
```

```python
def calc_tf(corpus, all_words) :
  # Number of documents in the corpus
  n_docs = len(corpus)
  # Number of unique words in the corpus
  n_unique_words = len(all_words)

  df_tf = pd.DataFrame(np.zeros((n_docs, n_unique_words)), columns=all_words)

  # Compute Term Frequency (TF)
  for i in range(n_docs):
      words = corpus[i].split()
      for w in words:
          df_tf.loc[i,w] = df_tf.loc[i,w] + (1 / len(words))
  return df_tf
```

```python
df_tf = calc_tf(corpus, list(all_words))
df_tf.head()
```

```python
def calc_idf(corpus, all_words) :
  idf = {}
  n_docs = len(corpus)
  for w in all_words:
      k = 0    # number of documents in the corpus that contain this word

      for i in range(n_docs):
          if w in corpus[i].split():
              k += 1

      idf[w] =  np.log((n_docs+1) / (k+1)) +1
  return idf
```

```python
idf = calc_idf(corpus, list(all_words))
print(idf)
```

```python
def calc_tf_idf(corpus, all_words):
  # Calculate TF
  df_tf = calc_tf(corpus, all_words)
  # Calculate IDF
  idf = calc_idf(corpus, all_words)
  df_tf_idf = df_tf.copy()
  # Caculate TF-IDF
  for w in all_words:
    for i in range(len(corpus)):
        df_tf_idf.loc[i,w] = df_tf.loc[i,w] * idf[w]
  # Calculate Normalized TF-IDF
  for i in range(len(corpus)):
    root_sum_sq = np.sqrt(np.sum(np.square(df_tf_idf.loc[i,:])))
    for w in all_words:
        df_tf_idf.loc[i,w] = df_tf_idf.loc[i,w]/ root_sum_sq
  return df_tf_idf
```

```python
tf_idf = calc_tf_idf(corpus, list(all_words))
# Sort alphabeticaly
tf_idf = tf_idf.reindex(sorted(tf_idf.columns), axis=1)
tf_idf.head()
```
